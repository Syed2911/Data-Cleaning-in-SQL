Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

Data collection and Processing

```python
# loading the csv data to a pandas dataframe
gold_data = pd.read_csv('/content/gld_price_data.csv')
```

```python
# print first 5 rows in DB
gold_data.head()
```

```python
# print last 5 rows of dataframe name
gold_data.tail()
```

```python
# number of rows and columns
gold_data.shape
```

```python
# getting some basic info abt data
gold_data.info()
```

```python
# checking the number of missing values
gold_data.isnull().sum()
```

```python
# getting stats measure of data
gold_data.describe()
```

Find corelation between 2 data
Corelation
1. Positive corelation
2. Negative corelation

```python
gold_data = gold_data.drop('Date',axis=1)
correlation = gold_data.corr()
```

```python
# comstructing a heatmap to understand the coorelation
plt.figure(figsize = (8,8))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
# correlation avlues of GLD
print (correlation['GLD'])
```

```python
# check the distribution of gold price
sns.distplot(gold_data['GLD'],color='green')
```

Splitting the feature and Target

```python
X = gold_data.drop(['GLD'],axis=1)
Y = gold_data['GLD']
```

```python
print(X)
```

```python
print (Y)
```

Splitting into traing and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

Model training
Random forest regression

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
# tarin the model
regressor.fit(X_train,Y_train)
```

Model evaluation

```python
# prediction on test data
test_data_prediction = regressor.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
# R squared error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print ("R squared error : ", error_score)
```

Compare actual values and predicted value in plot

```python
Y_test = list(Y_test)
```

```python
plt.plot(Y_test, color='blue', label = 'Actual Value')
plt.plot(test_data_prediction, color='green', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
